```python
# Import Modules

import pandas as pd
import os

from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
from pathlib import Path
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler

# Display all of the columns
pd.set_option('display.max_columns', None)
```

```python
#Read the CSV file into a Pandas DataFrame
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Proj4/train.csv")
df
```

```python
#Data cleaning
#Drop unnecessary columns
columns_to_drop = ["ID", "CustomerID", "Month", "Name", "Age",
                   "SSN", "Occupation", "TypeofLoan", "InterestRate", "ChangedCreditLimit","MonthlyInhandSalary",
                   "PaymentofMinAmount", "TotalEMIpermonth", "Amountinvestedmonthly",
                   "PaymentBehaviour"]
cleandf = df.drop(columns = columns_to_drop)
cleandf
```

```python
#Convert CreditHistoryAge string data value to usable form by extracting numerical part
cleandf['CreditHistoryAge'] = cleandf['CreditHistoryAge'].str[:2]
cleandf
```

```python
#Drop Creditmix rows with blank
cleandf = cleandf.dropna(subset=['CreditMix'])
```

```python
#Converting String to Int
cleandf['CreditHistoryAge'] = pd.to_numeric(cleandf['CreditHistoryAge'], errors='coerce')
```

```python
#Binning Target Column
bin_mapping = {
    'Good': 'Good/Standard',
    'Standard': 'Good/Standard',
    'Poor': 'Poor'
}

# Create a new column for the binned categories
cleandf['target'] = cleandf['CreditScore'].map(bin_mapping)
cleandf = cleandf.drop(columns = 'CreditScore')
cleandf
```

```python
#Check Datatypes
cleandf.dtypes
```

```python
#Fill in blank field with zeros
cleandf = cleandf.fillna(0)
cleandf
```

```python
#Export clean data to csv
csvpath = "/content/drive/MyDrive/Colab Notebooks/Proj4/traincleanresult.csv"
cleandf.to_csv(csvpath, index=False)
```

```python
#Check data types
cleandf.dtypes
```

```python
# Split target column from dataset
y = cleandf['target']
X= cleandf.drop(columns='target')
```

```python
# Preview the data
X[:8]
```

```python
# Print first ten entries for target
y[:10]
```

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
# Preview the data
X.head()
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#Check shape of X_train
X_train.shape
```

```python
# Check shape of X_test
X_test.shape
```

```python
# Create the StandardScaler instance
scaler = StandardScaler()
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)
# Scale the training data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_train_scaled
```

```python
# Instantiate the model with k = 3 neighbors
model = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model
model.fit(X_train_scaled, y_train)
```

```python
# Create predictions
y_pred = model.predict(X_test_scaled)

# Review the predictions
y_pred
```

```python
# Print confusion matrix
confusion_matrix(y_pred,y_test)
```

```python
# Print classification report
print(classification_report(y_pred,y_test))
```
