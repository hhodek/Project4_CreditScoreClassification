**Cleaning Data**

```python
train_df= train_df.fillna(0)
train_df.head()
```

```python
# Trying to remove values from the "Payment_Behavior" column that do not have any weight in the data
train_df = train_df[train_df["Payment_Behaviour"] != "!@9#%8"]

train_df.head(2)
```

```python
#Removing any Null values and replacing them with zero
train_df = train_df.fillna(0)
train_df = train_df.dropna()
train_df.head()
```

```python
# Extract the first two characters from the "Credit_History_Age" column
train_df["Credit_History_Age"] = train_df["Credit_History_Age"].str[:2]

# Remove rows where "Credit_History_Age" is '0'
train_df = train_df[train_df["Credit_History_Age"] != '0']

# Convert the remaining values to numeric, coercing non-numeric values to NaN
train_df["Credit_History_Age"] = pd.to_numeric(train_df["Credit_History_Age"], errors="coerce")

# Display the first few rows of the DataFrame
train_df.head()
```

```python
train_df["Monthly_Balance"] = pd.to_numeric(train_df["Monthly_Balance"], errors ="coerce")
train_df["Monthly_Balance"] = train_df["Monthly_Balance"].round(2)
```

```python
train_df["Outstanding_Debt"] = train_df["Outstanding_Debt"].astype(str)
train_df = train_df[~train_df["Outstanding_Debt"].str.endswith('_')]
train_df["Outstanding_Debt"] = train_df["Outstanding_Debt"].astype(float)

train_df.head(2)
```

```python
train_df= train_df.dropna(subset=['Credit_Score'])
train_df = train_df[train_df["Credit_Score"] != '0']
```

```python
#Wanted to make the percentages in the "Credit_Utilization_Ratio" column into decimal points rounded to the nearest thousandths place
scaling_factor = 0.01
train_df["Credit_Utilization_Ratio"] = train_df["Credit_Utilization_Ratio"] * scaling_factor
train_df["Credit_Utilization_Ratio"] = train_df["Credit_Utilization_Ratio"].round(3)
train_df.head(3)
```

```python
# Binned the values in the "Payment_Behavior" columns to simplify them a bit more. 
# I focused more on the payments made rather than the spending 
binning = {
    "Low_spent_Small_value_payments" : "Small_payments",   
    "High_spent_Medium_value_payments" : "Medium_payments",   
    "Low_spent_Medium_value_payments" : "Medium_payments",   
    "High_spent_Large_value_payments" : "Large_payments",   
    "High_spent_Small_value_payments" : "Small_payments",   
    "Low_spent_Large_value_payments" : "Large_payments"    
}

train_df["Payment_Behaviour"] = train_df["Payment_Behaviour"].map(binning)
train_df.head(3)
```

```python
binner = {
    "Good" : "Good/Standard",
    "Standard" : "Good/Standard",
    "Poor" : "Poor"
}
train_df["Credit_Score"] = train_df["Credit_Score"].map(binner)
train_df.head()
```

```python
train_df= train_df.dropna(subset=['Credit_Mix'])
train_df = train_df[train_df["Credit_Mix"] != '0']
```

```python
#Binning CreditMix Column
bin_map = {
    'Good': 'Good',
    'Standard': 'Good',
    'Bad': 'Bad'
}

# Cover old column with the binned categories
train_df['Credit_Mix'] = train_df['Credit_Mix'].map(bin_map)

train_df
```

```python
#Creating a set of columns to drop from the dataframe that We do not believe will effect our analysis
delete_them = ["ID", "SSN", "Customer_ID", "Payment_of_Min_Amount", "Total_EMI_per_month", 
               "Monthly_Inhand_Salary", "Delay_from_due_date",
               "Age", "Amount_invested_monthly", "Changed_Credit_Limit", "Num_of_Delayed_Payment", 
               "Interest_Rate", "Type_of_Loan"]
```

```python
# Dropping the columns using the variable above
train_df = train_df.drop(columns= delete_them)
train_df.head(3)
```

```python
train_df= train_df.dropna()
train_df.head()
```

```python
train_df = train_df[['Name', 'Occupation', 'Month', 
       'Payment_Behaviour', 'Credit_Score', 'Credit_Mix',
       'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Credit_History_Age',
        'Num_Credit_Card', 'Monthly_Balance']]
train_df.head()
```
